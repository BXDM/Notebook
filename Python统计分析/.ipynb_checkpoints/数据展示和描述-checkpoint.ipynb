```python
# 2.3 众数
from scipy import stats
data = [1, 3, 4, 4, 7]
```

```python
x = np.arange(1, 101)
```

```python
# 2.4 几何均值
stats.gmean(x)
```

**算术均值的计算**

$ mean = \frac{\sum}{n}  $

**几何均值的计算**

$$ mean_{geometric} = \sqrt[N]{\prod_{i = 1}^{N} x_i} = \exp \left( \frac{\sum_i \ln(x_i)} {n} \right) $$
