```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
s = pd.DataFrame(np.random.randn(1000)+10, columns = ['value'])
print(s.head())
fig = plt.figure(figsize = (12, 6)) 

plt.subplot(311)
plt.scatter(s.index, s.values)
plt.grid()

plt.subplot(312)
plt.hist(s.values, bins = 30, alpha = 0.5, color = 'orange')
plt.grid()

# 使用 Seaborn 绘制密度曲线
plt.subplot(313)
import seaborn as sns
sns.distplot(s.values, hist = False, rug = 'True')

# 存储图片
plt.savefig('正太分布.png', dpi = 400, bbox_inches = 'tight')  # bbox_inches 将图片完整保存
```

```python
s = pd.DataFrame(np.random.randn(1000)+10,columns = ['value'])
print(s.head())
# 创建随机数据

fig = plt.figure(figsize = (10,6))
ax1 = fig.add_subplot(2,1,1)  # 创建子图1
ax1.scatter(s.index, s.values)
plt.grid()
# 绘制数据分布图

ax2 = fig.add_subplot(2,1,2)  # 创建子图2
s.hist(bins=30,alpha = 0.5,ax = ax2)
s.plot(kind = 'kde', secondary_y=True,ax = ax2)
plt.grid()
# 绘制直方图
# 呈现较明显的正太性
```
