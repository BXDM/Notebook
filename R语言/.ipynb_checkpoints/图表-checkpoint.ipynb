### R 语言绘图包
- graphics
- grid
- lattice

```r
# 导入 nutshell
library(nutshell)
data(field.goals)
```

```r
# 查看数据框列名
names(field.goals)
```

```r
hist(field.goals$yards)
```

```r
hist(field.goals$yards, breaks = 35)
# break 添加组数
```

```r
table(field.goals$play.type)
```

```r
stripchart(field.goals[field.goals$play.type == "FG blocked", ]$yards,
          pch = 19, method = "jitter")
```

```r
plot(cars, xlab = 'speed(mph)', ylab = "stopping distance(ft)",
    las = 1, xlim = c(0, 25))
```

```r
# 载入 lattice 图形
library(lattice)
data(consumption)
```

```r
dotplot(Amount~Year | Food, data = consumption, 
       aspect = 'xy', scale = list(relation = "sliced", cex = 0.4))
# aspect 选项可以改变图片的纵横比
```
