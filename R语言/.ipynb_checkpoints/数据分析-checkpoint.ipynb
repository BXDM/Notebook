```r
library(nutshell)
```

```r
data(dow30)
```

```r
dow30
```

```r
mean(dow30$Open)
```

```r
min(dow30$Open)
```

```r
max(dow30$Open)
```

```r
mean(c(1, 2, 3, 4, 5, NA))
```

```r
# 使用 na.rm 参数移除缺失值
mean(c(1, 2, 3, 4, 5, NA), na.rm = TRUE)
```

```r
mean(c(-1, 0:100, 2000))
```

```r
# 使用 trim  函数处理极端值
mean(c(-1, 0:100, 2000), trim  = 0.1)
```

```r
# range 函数返回最大值和最小值
range(dow30$Open)
```

```r
# quantitle 函数返回分位数
quantile(dow30$Open, probs = c(0, 0.25, 0.5, 0.75, 1.0))
```

```r
# IQR 计算四分位距
IQR(dow30$Open)
```
