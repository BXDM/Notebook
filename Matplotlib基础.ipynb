#### 使用 `facecolor` 参数设置画布颜色
参考资料：<https://www.cnblogs.com/jason--/p/11456520.html>

````python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 4 * np.pi, 1000)
r = 0.8 * theta
plt.figure("Polar", facecolor='lightgray')
plt.gca(projection='polar')
plt.title('Polar', fontsize=16)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$\rho$')
plt.grid(linestyle=':')
plt.plot(theta, r)
x = np.linspace(0, 6 * np.pi, 1000)
y = 3 * np.sin(6 * x)
plt.plot(x, y)
plt.show()
````

## 绘制子图

````python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec

fig = plt.figure(tight_layout=True)
gs = gridspec.GridSpec(2, 2)

ax = fig.add_subplot(gs[0, :])
ax.plot(np.arange(0, 1e6, 1000))
ax.set_ylabel('YLabel0')
ax.set_xlabel('XLabel0')

for i in range(2):
    ax = fig.add_subplot(gs[1, i])
    ax.plot(np.arange(1., 0., -0.1) * 2000., np.arange(1., 0., -0.1))
    ax.set_ylabel('YLabel1 %d' % i)
    ax.set_xlabel('XLabel1 %d' % i)
    if i == 0:
        for tick in ax.get_xticklabels():
            tick.set_rotation(55)
fig.align_labels()  # same as fig.align_xlabels(); fig.align_ylabels()

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt


def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)


t1 = np.arange(0.0, 3.0, 0.01)

ax1 = plt.subplot(212)
ax1.margins(0.05)           # Default margin is 0.05, value 0 means fit
ax1.plot(t1, f(t1))

ax2 = plt.subplot(221)
ax2.margins(2, 2)           # Values >0.0 zoom out
ax2.plot(t1, f(t1))
ax2.set_title('Zoomed out')

ax3 = plt.subplot(222)
ax3.margins(x=0, y=-0.25)   # Values in (-0.5, 0.0) zooms in to center
ax3.plot(t1, f(t1))
ax3.set_title('Zoomed in')

plt.show()
````

## 设置坐标轴刻度

#### 使用 xlim 和 ylim 设置坐标的刻度范围

````python
plt.plot(x, x, label = 'liner')
plt.plot(x, x**2, label = 'quadratic')
plt.plot(x, x**3, label = 'cubic')

plt.xlim(0,1.25)
plt.ylim(0,2)
````

#### 设置对数坐标
semilogx——设置 x 为对数坐标

semilogy——设置 y 为对数坐标

loglog——设置 xy 为对数坐标

````python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (12, 4))
plt.subplot(131)
plt.semilogx()
plt.subplot(132)
plt.semilogy()
plt.subplot(133)
plt.loglog()
````

## 坐标轴标签倾斜

使用 `xticks(rotation = degree)` 属性

degree —— 旋转角度

参考资料：[How to rotate date ticks using Matplotlib in Python](https://www.kite.com/python/answers/how-to-rotate-date-ticks-using-matplotlib-in-python)

````python
import numpy as np
x = np.linspace(0, 2, 100)
plt.plot(x, x, label = 'liner')
plt.plot(x, x**2, label = 'quadratic')
plt.plot(x, x**3, label = 'cubic')
plt.xticks(rotation = 30)
````
