```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# 获取近两年人民币兑美元汇率数据
ticker = 'CNY=X'
data = yf.download(ticker, period="2y", interval="1d")

# 只保留收盘价
data = data[['Close']]

# 找出两次最低点
min_points = data.nsmallest(2, 'Close')

# 绘制汇率曲线图
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Close'], label='CNY to USD Exchange Rate')

# 画出两次最低点的水平线
for idx in min_points.index:
    plt.axhline(y=min_points.loc[idx]['Close'], color='red', linestyle='--', label=f'Lowest Point: {min_points.loc[idx]["Close"]:.4f}')

# 添加标题和标签
plt.title('CNY to USD Exchange Rate - Last 2 Years')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.legend()
plt.grid(True)

# 显示图表
plt.show()
```
