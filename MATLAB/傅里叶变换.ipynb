#### 分析受噪声干扰时的信号

```matlab
clear all;
Fs = 1000;
T = 1/Fs;
L = 1000;
t = (0:L-1)*T;
x = 0.7*sin(2*pi*50*t) + sin(2*pi*120*t);
y = x+2*randn(size(t));
plot(Fs*t(1:50), y(1:50))
title('受噪声干扰的信号')
xlabel('时间/毫秒')
ylabel('频率')
```

转变到时间域

```matlab
NFFT = 2^nextpow2(L);
Y = fft(y, NFFT)/L;
f = Fs/2*linspace(0, 1, NFFT/2+1);
plot(f, 2*abs(Y(1:NFFT/2+1)))
title('y的频率成分')
xlabel('频率/Hz')
ylabel('|Y(f)|')
```
