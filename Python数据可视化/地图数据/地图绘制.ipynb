```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Read geographic data from the local shapefile
world = gpd.read_file('ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp')

# Check the columns in the dataset
print(world.columns)

# Plot basic map
world.plot()
plt.title("World Map")
plt.show()

# Plot map with country names and different colors for continents
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax)

# Assuming the correct column for continents is 'CONTINENT'
world.plot(column='CONTINENT', ax=ax, legend=True,
           legend_kwds={'title': "Continent"})

# Add country names
for x, y, label in zip(world.geometry.centroid.x, world.geometry.centroid.y, world.NAME):
    ax.text(x, y, label, fontsize=8)

plt.title("World Map with Continents and Country Names")
plt.show()
```
