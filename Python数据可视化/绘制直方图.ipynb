```python
from numpy.random import randn
import matplotlib.pyplot as plt
```

```python
plt.hist(randn(100), bins = 20, alpha = 0.4, edgecolor = 'white')
```

hist 方法绘制直方图
- x 绘图使用的参数
- bins 条的数量
- alpha 透明度
- edgecolor 边框颜色
