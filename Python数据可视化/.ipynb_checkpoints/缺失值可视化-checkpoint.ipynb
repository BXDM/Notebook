```python
# 需要先安装 missingno 模块
import missingno as msno
import pandas as pd
import numpy as np
from numpy import nan as NA
```

```python
# 创建 DataFrame
df = pd.DataFrame(np.random.randn(8, 10))
df[df<0] = NA
df
```

```python
msno.matrix(df, labels = True)
```

```python
msno.bar(df)
# 查看每列的缺失比例
```

```python
msno.heatmap(df)
# 相关性热图
```

```python
msno.dendrogram(df)
```
